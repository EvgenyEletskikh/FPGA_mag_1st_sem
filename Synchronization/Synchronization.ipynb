## Моделирование беспроводного канала

Прежде чем мы узнаем, как реализовать временную и частотную синхронизацию, приведем наши смоделированные сигналы к более реалистичному виду. Без добавления некоторой случайной временной задержки процесс синхронизации во времени тривиален. В таком случае вам нужно только учесть в схеме задержку всех используемых фильтров. Также добавим в схему частотное смещение, потому что, как мы обсудим далее, генераторы не идеальны и между центральной частотой передатчика и приемника всегда будет некоторое смещение.

Давайте рассмотрим код Python для имитации нецелочисленной задержки и смещения частоты. Код Python в этой главе будет начинаться с кода, который мы написали во время упражнения Python по формированию импульсов; вы можете считать его отправной точкой кода в этой главе, а весь новый код будет следовать за ним.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import math

# это часть кода из упражнения по генерации импульса
num_symbols = 100
sps = 8 # количество отсчетов на символ
bits = np.random.randint(0, 2, num_symbols) # Данные на выходе передатчика, нули и единицы
pulse_train = np.array([])
for bit in bits:
    pulse = np.zeros(sps)
    pulse[0] = bit*2-1 # устанавливаем значение первого элемента равным 1 или -1
    pulse_train = np.concatenate((pulse_train, pulse)) # добавляем к сигналу 8 отсчетов

# Генерация фильтра приподнятого косинуса
num_taps = 101
beta = 0.35
Ts = sps # Примем частоту дискретизации равной 1 Гц, тогда период отсчетов равен 1, а период символов - 8
t = np.arange(-51, 52) # помните, что последнее число не входит в массив
h = np.sinc(t/Ts) * np.cos(np.pi*beta*t/Ts) / (1 - (2*beta*t/Ts)**2)

# Отфильтруем наш сигнал, чтобы применить формирование импульса
samples = np.convolve(pulse_train, h)
````

Мы опустим код, связанный с построением графика, потому что к настоящему моменту вы, вероятно, научились отображать любой сигнал, который хотите. Чтобы графики выглядели красиво, как это часто делается в этом учебнике, требуется много дополнительного кода, который не обязательно понимать.

## Добавление задержки

Мы легко можем имитировать задержку, сдвигая отсчеты, но так можно получить только целочисленную задержку, кратную нашему периоду дискретизации. В реальном мире задержка будет составлять некоторую долю периода дискретизации. Мы можем смоделировать задержку части отсчета, создав фильтр _«дробной задержки»_, который пропускает все частоты, но задерживает отсчеты на некоторую величину, не ограниченную шагом дискретизации. Вы можете думать об этом как о всепропускающем фильтре (_all-pass filter_), который применяет одинаковый фазовый сдвиг ко всем частотам (напомним, что временная задержка и фазовый сдвиг эквивалентны). Код Python для создания этого фильтра показан ниже:

````python
# Генерируем и применяем фильтр "дробной задержки"
delay = 0.4 # дробная задержка, указанная в количестве отсчетов
N = 21 # количество отсчетов фильтра
n = np.arange(-N//2, N//2) # ...-3,-2,-1,0,1,2,3...
h = np.sinc(n - delay) # получаем значения отсчетов для фильтра
h *= np.hamming(N) # применяем к фильтру оконную функцию, чтобы убедиться, что фильтрующая функция сходится к нулю в обе стороны
h /= np.sum(h) # нормирование отсчетов, чтобы получить единичное усиление на выходе фильтра, т.к. мы не хотим изменять амплитуду/мощность сигнала
samples_f = np.convolve(samples, h) # применяем фильтр
````

Как видите, мы вычисляем отсчеты фильтра с помощью функции `sinc()`. _Sinc_ во временной области — это прямоугольник в частотной области, и наш прямоугольник для этого фильтра охватывает весь частотный диапазон нашего сигнала. Этот фильтр не изменяет форму сигнала, а просто задерживает его во времени. В нашем примере мы запаздываем на **0,4** отсчета. Имейте в виду, что применение любого фильтра задерживает сигнал на половину отсчетов фильтра минус один из-за свертки сигнала через фильтр.

Если мы построим график «до» и «после» фильтрации сигнала, мы можем наблюдать дробную задержку. В нашем сюжете мы приближаем только пару символов. В противном случае дробная задержка не отображается.

````python
#FIXME: удалить ячейку и вставить картинку

plt.figure()

a = 50
b = 200
plt.plot(samples[a:b],'.-')
plt.plot(samples_f[a:b], '.-', color='Orange')
plt.legend(["До", "После"])
plt.xlabel("Отсчеты")
plt.ylabel("Действительная часть сигнала");
````

## Добавление частотного смещения

Чтобы сделать наш моделируемый сигнал более реалистичным, мы внесем частотную рассинхронизацию. Предположим, что наша частота дискретизации составляет 1 МГц (на самом деле не важно, чему равна частота дискретизации, но вы увидите, почему мы выбрали именно такое значение). Если мы хотим смоделировать смещение частоты на 13 кГц (какое-то произвольное число), мы можем сделать это с помощью следующего кода:

````python
# применение частотного рассогласования к сигналу
fs = 1e6 # примем частоту дискретизации равной 1 МГц
fo = 13000 # величина частотного смещения, Гц
Ts = 1/fs # период дискретизации
t = np.arange(0, Ts*len(samples_f), Ts)
samples_shifted = samples_f * np.exp(1j*2*np.pi*fo*t) # применяем частотное рассогласование
````

## Временнáя синхронизация

Когда мы передаем сигнал по беспроводной сети, он достигает приемника со случайным фазовым сдвигом из-за времени, необходимого для распространения сигнала. Мы не можем просто начать считывать отсчеты символов с нашей символьной скоростью, потому что мы вряд ли будем производить выборку в нужном месте импульса, как обсуждалось в конце главы «[Формирование импульса](https://pysdr.org/content/pulse_shaping.html)». Изучите три рисунка в конце этой главы, если вы не понимаете.

Большинство методов временной синхронизации использую принцип работы контура фазовой [автоподстройки частоты](https://ru.wikipedia.org/wiki/%D0%A4%D0%B0%D0%B7%D0%BE%D0%B2%D0%B0%D1%8F_%D0%B0%D0%B2%D1%82%D0%BE%D0%BF%D0%BE%D0%B4%D1%81%D1%82%D1%80%D0%BE%D0%B9%D0%BA%D0%B0_%D1%87%D0%B0%D1%81%D1%82%D0%BE%D1%82%D1%8B) (ФАПЧ или PLL); мы не будем изучать здесь PLL, но важно знать этот термин, и хотя бы в общих чертах понимать принцип работы. PLL — это системы с обратной связью, которые используют обратную связь для постоянной регулировки чего-либо; в нашем случае временной сдвиг позволяет нам производить выборку на пике цифровых символов.

Вы можете представить восстановление синхронизации как блок в приемнике, который принимает один поток отсчетов и выводит другой поток отсчетов (аналогично фильтру). Мы программируем этот блок с информацией о нашем сигнале, наиболее важной из которых является количество отсчетов на символ (или устанавливаем предположение о значении этой величины, если мы не уверены на 100%, что было передано). Этот блок действует как «дециматор», т. е. размер выборки на выходе будет частью размера входной выборки. Нам нужен один отсчет на цифровой символ, поэтому величина децимации просто будет равна количеству отсчетов на символ. Если передатчик передает со скоростью 1 млн символов в секунду, а мы делаем выборку со скоростью 16 млн выборок в секунду, мы получим 16 отсчетов на символ. Это будет частота дискретизации, поступающая в блок синхронизации времени. Частота дискретизации сигнала на выходе блока будет равна 1 Мвыб/с, потому что нам нужен один отсчет на цифровой символ.

Большинство методов временной синхронизации основаны на том факте, что наши цифровые символы "растут", а затем "падают", а вершина — это точка, в которой мы хотим произвести отсчет символа. Другими словами, мы выбираем максимальную точку после получения абсолютного значения:

# <КАРТИНКА>

Существует множество методов временной синхронизации, очень похожих на PLL. Как правило, разница между ними заключается в уравнении, используемом для выполнения «коррекции» временного смещения, которое мы обозначаем как $\mu$ или `mu` в коде. Значение `mu` выражается в единицах сэмплов м обновляется при каждой итерации цикла. Вы можете думать об этом как "_на сколько отсчетов мы должны сместиться, чтобы иметь возможность произвести отсчет в «идеальное» время_". Таким образом, если `mu = 3,61`, это означает, что мы должны сдвинуть входные данные на 3,61 отсчета, чтобы сэмплировать в нужном месте. Поскольку у нас 8 отсчетов на символ, если `mu` превысит 8, то `mu` просто вернется к нулю.

Следующий код Python реализует синхронизацию согласно [алгоритму Миллера и Мюллера](https://wirelesspi.com/mueller-and-muller-timing-synchronization-algorithm/).

````python
mu = 0 # исходная оценка фазы символа
out = np.zeros(len(samples_f) + 10, dtype=complex)
out_rail = np.zeros(len(samples_f) + 10, dtype=complex) # хранилище значений отсчетов; на каждой итерации нам необходимы два предыдущих значения плюс текущее
i_in = 0 # индекс входных отсчетов
i_out = 2 # индекс выходных отсчетов (пусть первые два отсчета будут нулями)
while i_out < len(samples) and i_in+16 < len(samples):
    out[i_out] = samples[i_in + int(mu)] # берем "лучший" отсчет
    out_rail[i_out] = int(np.real(out[i_out]) > 0) + 1j*int(np.imag(out[i_out]) > 0)
    x = (out_rail[i_out] - out_rail[i_out-2]) * np.conj(out[i_out-1])
    y = (out[i_out] - out[i_out-2]) * np.conj(out_rail[i_out-1])
    mm_val = np.real(y - x)
    mu += sps + 0.3*mm_val
    i_in += int(np.floor(mu)) # округляем вниз до целого значения, чтобы использовать его как индекс
    mu = mu - np.floor(mu) # избавляемся от целочисленной части
    i_out += 1 # инкрементируем индекс выходных отсчетов
out = out[2:i_out] # отрезаем первые два отсчета и все те, что идут после i_out
samples_MM = out # включите эту строку только в том случае, если вы хотите соединить этот фрагмент кода с петлей Костаса позже
````

В блок восстановления синхронизации подаются «принятые» отсчеты, и он генерирует выходные отсчеты по одному за раз (обратите внимание, что `i_out` увеличивается на 1 на каждой итерации цикла). Блок восстановления не выкидывает просто «принятые» отсчеты один за другим на выход благодаря тому, как в цикле изменяется `i_in`. Он будет пропускать некоторые сэмплы в попытке получить «правильный», который будет соответствовать пику импульса. Когда цикл обрабатывает сэмплы, он медленно синхронизируется с символом или, по крайней мере, пытается это сделать, настраивая `mu`. Согласно коду, целая часть `mu` добавляется к `i_in`, а затем вычитается из `mu` (имейте в виду, что `mm_val` может принимать как отрицательные так и положительным значения на каждой итерации цикла). После полной синхронизации цикл должен осуществлять выборку только в центре каждого символа/импульса. Вы можете настроить константу `0,3`, которая изменит чувствительность контура обратной связи; более высокое значение повысит чувствительность контура, но может привести к полной рассинхронизации.

## Синхронизация времени с интерполяцией

Символьные синхронизаторы имеют тенденцию интерполировать входные отсчеты на некоторое число, например, 16, чтобы можно было сместиться на долю выборки (т.е. каждый отсчет повторяется 16 раз). Случайная задержка, вызванная беспроводным каналом, вряд ли будет точно кратна выборке, поэтому отсчеты скорее всего будут производиться не в максимуме символа, а где-то рядом. Это особенно верно в случае, когда может быть только 2 или 4 выборки на принимаемый символ. Интерполяция входных отсчетов дает нам возможность работать «между» фактическими отсчетами, чтобы попасть в самый пик каждого символа. Выход синхронизатора по-прежнему составляет только 1 отсчет на символ. Сами входные отсчеты интерполируются.

Наш код Python для временной синхронизации, который мы реализовали выше, не включал никакой интерполяции. Чтобы расширить наш код, включите дробную временную задержку, которую мы реализовали в начале этой главы, чтобы наш полученный сигнал имел более реалистичную задержку. Пока оставьте смещение частоты отключенным. Если вы повторно запустите симуляцию, вы обнаружите, что синхронизатор не может полностью синхронизироваться с сигналом. Это потому, что мы не интерполируем, поэтому у кода нет возможности «отсчитывать между отсчетами», чтобы компенсировать дробную задержку. Добавим интерполяцию.

Быстрый способ интерполировать сигнал в Python — использовать `signal.resample` или `signal.resample_poly` из библиотеки `scipy`. Обе эти функции делают одно и то же, но работают по-разному внутри. Мы будем использовать последнюю функцию, потому что она работает быстрее. Давайте интерполируем на 16 (выбрано произвольно, вы можете попробовать разные значения), т. е. мы будем вставлять 15 дополнительных отсчетов между каждыми соседними отсчетами. Это можно сделать в одной строке кода, и это должно произойти до того, как мы перейдем к временной синхронизации (до большого фрагмента кода выше). Давайте также построим график до и после, чтобы увидеть разницу:

````python
samples_interpolated = signal.resample_poly(samples, 16, 1)

# Строим графики "до" и "после"
fig, [ax1, ax2] = plt.subplots(2,1)

ax1.set_title('До интерполяции')
ax1.plot(samples[0:4],'.-')

ax2.set_title('После интерполяции')
ax2.plot(samples_interpolated[0:64],'.-')

fig.tight_layout()
````

Из рисунка видно, что точек на графике стало гораздо больше, а именно - в 16 раз.

Надеюсь, теперь вам ясно, для чего нужна интерполяция сигнала. Эти дополнительные отсчеты позволят нам учесть дробную часть задержки снятия отсчетов. В дополнение к вычислению `sample_interpolated` нам также нужно изменить одну строку кода в нашем временном синхронизаторе. Мы изменим первую строку внутри цикла `while` на следующую:
